```
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import StackingClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection
```

```
# Load a dataset into a Pandas Dataframe.
df = pd.read_csv('/work/Dataset1.csv')

# Display the first 3 examples.
df.head(3)
```

```
columns_to_drop = ['He leído el consentimiento y las explicaciones del investigador, por lo que reconozco que al completar y entregar esta escala estoy dando mi consentimiento para participar en esta investigación .',
                    'No podía sentir ningún sentimiento positivo',
                    'Me di cuenta que tenia la boca seca',
                    'Me costó mucho relajarme',
                    'Se me hizo difícil tomar la iniciativa para hacer cosas',
                    'Sentí que no tenia nada por que vivir',
                    'Me sentí triste y deprimido',
                    'No me pude entusiasmar por nada',
                    'Sentí que valía muy poco como persona',
                    'Sentí que la vida no tenia ningún sentido',
                    'Se me hizo difícil respirar',
                    'Sentí que mis manos temblaban',
                    'Estaba preocupado por situaciones en las cuales podía tener pánico o en las que podría hacer el ridículo',
                    'Sentí que estaba al punto de pánico',
                    'Sentí los latidos de mi corazón a pesar de no haber hecho ningún esfuerzo físico',
                    'Tuve miedo sin razón',
                    'Reaccioné exageradamente en ciertas situaciones',
                    'Sentí que tenia muchos nervios',
                    'Noté que me agitaba',
                    'Se me hizo difícil relajarme',
                    'No toleré nada que no me permitiera continuar con lo que estaba haciendo',
                    'Sentí que estaba muy irritable',
                    'Porcentaje Violencia',
                    'Depresión',
                    'Ansiedad',
                    'Estres'
                    ]
df3 = df.drop(columns_to_drop, axis=1)
df3
```

```
# Variables categoricas
# Selecciono las columnas que son categoricas
# Aplic dummies para convertir variables categoricas
categorical_columns = ['Lugar de Procedencia', 
                        'Estado civil', 
                        'Nivel de estudios',
                        '1.- En general, ¿como describiría usted su relación de pareja?', 
                        '2.- ¿Con cuanta dificultad usted y su pareja resuelven sus diferencias?', 
                        '3.- ¿Al terminar las discusiones usted se siente decaída o mal con usted misma?', 
                        '4.- ¿Las discusiones terminan en golpes, patadas o empujones?', 
                        '5.- ¿Usted se siente aterrada por lo que su pareja pueda decir o hacer?', 
                        '6.- ¿Alguna vez su pareja ha abusado fisicamente de usted?', 
                        '7.- ¿Alguna vez su pareja ha abusado emocionalmente de usted?',
                        '8.- ¿Alguna vez su pareja ha abusado sexualmente de usted?',
                        'Actualmente, ¿recibe atención psicológica y/o psiquiátrica?',
                        'Actualmente, ¿cuenta con apoyo de familiares y/o amigos?',
                        'Departamento',
                        'Violencia'
                        ]
label_encoder = LabelEncoder()
 #Iterar a través de las columnas categóricas y aplicar label encoding a cada una
for column in categorical_columns:
    df3[column] = label_encoder.fit_transform(df3[column])

     # Obtener la asignación de números a valores únicos
    mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    print(f'Mapping para la columna {column}: {mapping}')
df3.head(20)

#df_encode = pd.get_dummies(data = df, columns = categorical_columns,
             # drop_first =True)
#df_encode.head(100)
```

```
# Genero datos de entrenamiento y test
X3 = df3.drop('Violencia',axis=1) # Variable independiente
y3 = df3['Violencia'] # Variable dependiente

X_train3, X_test3, y_train3, y_test3 = train_test_split(X3,y3,test_size=0.2)
```

```
clf1 = DecisionTreeClassifier()
clf2 = DecisionTreeClassifier(criterion="gini", splitter="best", max_features=20, ccp_alpha=0.01)
clf3 = DecisionTreeClassifier(criterion="gini", splitter="random", max_features=30, ccp_alpha=0.01)
clf4 = DecisionTreeClassifier(criterion="entropy", splitter="random", max_features=10, ccp_alpha=0.01)
clf5 = DecisionTreeClassifier(criterion="entropy", splitter="best", max_features=20, ccp_alpha=0.01)
clf6 = DecisionTreeClassifier(criterion="log_loss", splitter="best", max_features=30, ccp_alpha=0.01)
clf7 = DecisionTreeClassifier(criterion="log_loss", splitter="random", max_features=10, ccp_alpha=0.01)
```

```
mmclf = LogisticRegression()
```

```
stacking_clf = StackingClassifier(estimators=[
    ('clf1', clf1),
    ('clf2', clf2),
    ('clf3', clf3),
    ('clf4', clf4),
    ('clf5', clf5),
    ('clf6', clf6),
    ('clf7', clf7)
], final_estimator=mmclf)
```

```
stacking_clf.fit(X_train3, y_train3)
clf1.fit(X_train3, y_train3)
clf2.fit(X_train3, y_train3)
clf3.fit(X_train3, y_train3)
clf4.fit(X_train3, y_train3)
clf5.fit(X_train3, y_train3)
clf6.fit(X_train3, y_train3)
clf7.fit(X_train3, y_train3)
```

```
predictionsclf1 = clf1.predict(X_test3)
predictionsclf2 = clf2.predict(X_test3)
predictionsclf3 = clf3.predict(X_test3)
predictionsclf4 = clf4.predict(X_test3)
predictionsclf5 = clf5.predict(X_test3)
predictionsclf6 = clf6.predict(X_test3)
predictionsclf7 = clf7.predict(X_test3)
predictions = stacking_clf.predict(X_test3)
predictions
```

```
accuracy = accuracy_score(y_test3, predictions)
accuracyclf1 = accuracy_score(y_test3, predictionsclf1)
accuracyclf2 = accuracy_score(y_test3, predictionsclf2)
accuracyclf3 = accuracy_score(y_test3, predictionsclf3)
accuracyclf4 = accuracy_score(y_test3, predictionsclf4)
accuracyclf5 = accuracy_score(y_test3, predictionsclf5)
accuracyclf6 = accuracy_score(y_test3, predictionsclf6)
accuracyclf7 = accuracy_score(y_test3, predictionsclf7)

print(f'Accuracy Staking: {accuracy}')
print(f'Accuracy clf1: {accuracyclf1}')
print(f'Accuracy clf2: {accuracyclf2}')
print(f'Accuracy clf3: {accuracyclf3}')
print(f'Accuracy clf4: {accuracyclf4}')
print(f'Accuracy clf5: {accuracyclf5}')
print(f'Accuracy clf6: {accuracyclf6}')
print(f'Accuracy clf7: {accuracyclf7}')
```
