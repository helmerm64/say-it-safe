```
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn import model_selection 
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, StratifiedKFold, cross_val_predict
from sklearn.feature_selection import SelectFromModel
from imblearn.over_sampling import RandomOverSampler


from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import recall_score, precision_score, f1_score, accuracy_score, confusion_matrix, classification_report
```

# Carga y procesamiento de datos

```
# Load a dataset into a Pandas Dataframe.
df = pd.read_csv('/work/Dataset2.csv')

# Display the first 3 examples.
df.head(3)
```

```
# Variables categoricas
# Selecciono las columnas que son categoricas
# Aplic dummies para convertir variables categoricas
categorical_columns = ['Lugar de Procedencia', 
                        'Estado civil', 
                        'Nivel de estudios',
                        'Actualmente, ¿recibe atención psicológica y/o psiquiátrica?',
                        'Actualmente, ¿cuenta con apoyo de familiares y/o amigos?',
                        'Departamento',
                        'Violencia'
                        ]
label_encoder = LabelEncoder()
 #Iterar a través de las columnas categóricas y aplicar label encoding a cada una
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

     # Obtener la asignación de números a valores únicos
    mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    print(f'Mapping para la columna {column}: {mapping}')
df.head(20)

#df_encode = pd.get_dummies(data = df, columns = categorical_columns,
             # drop_first =True)
#df_encode.head(100)
```

```
columns_to_drop = ['He leído el consentimiento y las explicaciones del investigador, por lo que reconozco que al completar y entregar esta escala estoy dando mi consentimiento para participar en esta investigación .',
                    '1.- En general, ¿como describiría usted su relación de pareja?', 
                    '2.- ¿Con cuanta dificultad usted y su pareja resuelven sus diferencias?', 
                    '3.- ¿Al terminar las discusiones usted se siente decaída o mal con usted misma?', 
                    '4.- ¿Las discusiones terminan en golpes, patadas o empujones?', 
                    '5.- ¿Usted se siente aterrada por lo que su pareja pueda decir o hacer?', 
                    '6.- ¿Alguna vez su pareja ha abusado fisicamente de usted?', 
                    '7.- ¿Alguna vez su pareja ha abusado emocionalmente de usted?',
                    '8.- ¿Alguna vez su pareja ha abusado sexualmente de usted?',
                    'Porcentaje Violencia',
                    'Depresion - Porcentaje',
                    'Ansiedad - Porcentaje',
                    'Estres - Porcentaje',
                    'Depresión',
                    'Ansiedad',
                    'Estres',
                    ]
df = df.drop(columns_to_drop, axis=1)
df
```

```
X = df.drop('Violencia',axis=1) # Variable independiente
y = df['Violencia'] # Variable dependiente
```

## Selección de características

Se realiza la selección de características utilizando importancia de características

Random Forest es comúnmente utilizado para este propósito debido a su capacidad para calcular la importancia de las características.

```
clf_feature_selection = RandomForestClassifier(n_estimators=100, random_state=42)
sfm = SelectFromModel(clf_feature_selection)
X_selected = sfm.fit_transform(X, y)
selected_features = X.columns[sfm.get_support()]
selected_features
```

```
feature_importances = sfm.estimator_.feature_importances_

# Calcular el porcentaje de importancia para cada característica
total_importance = sum(feature_importances)
percentage_importances = (feature_importances / total_importance) * 100

# Crear un DataFrame para mostrar las características y sus porcentajes de importancia
importance_df = pd.DataFrame({
    'Característica': X.columns,
    'Importancia (%)': percentage_importances
})

# Ordenar el DataFrame por importancia en orden descendente
importance_df = importance_df.sort_values(by='Importancia (%)', ascending=False)

# Mostrar el DataFrame
print(importance_df)
```

## Entrenamiento y ajuste de hiperparámetros de modelo

Entrenamiento y ajuste de hiperparámetros utilizando Grid Search con validación cruzada de 3 pliegues

En este paso utilizamos Grid Search para ajustar los hiperparámetros del árbol de decisión. Se buscó la profundidad del árbol (max_depth), el número mínimo de muestras por hoja (min_samples_leaf), el número mínimo de muestras necesarias para realizar una división (min_samples_split) y la función de medición de calidad del split (criterion). También medimos el valor de la AUC (área bajo la curva ROC) durante la validación cruzada de 3 pliegues

```
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
dt = DecisionTreeClassifier(random_state=42)
parameters = {'max_depth': [5, 10, 15, 20],
'min_samples_leaf': [1, 2, 3, 4, 5],
'min_samples_split': [2, 3, 4, 5],
'criterion': ['entropy', 'gini']
}
clf = GridSearchCV(dt, parameters, scoring='roc_auc', cv=skf, n_jobs=-1, verbose=1)
clf.fit(X_selected, y)
print("Mejores hiperparámetros: ", clf.best_params_)
print("Mejor valor de AUC: ", clf.best_score_)
```

### Over-sampling

En este paso utilizamos la técnica de sobre-muestreo (over-sampling) para tratar el desbalanceo de clases en el conjunto de datos. En este caso, utilizamos RandomOverSampler para equilibrar las clases minoritarias 

```
#Utilizamos RandomOverSampler de la bibiloteca imblearn
oversample = RandomOverSampler(sampling_strategy='minority')
X_oversampled, y_oversampled = oversample.fit_resample(X_selected, y)
```

## Construcción del modelo de Stack Ensemble

Se construyé el modelo de conjunto de árboles de decisión (stack ensemble) utilizando los parámetros ajustados en el paso 4 para el clasificador base. Se utilizaron 2 clasificadores base: el sencillo, sin ajustar sus parámetros y el mejor encontrado. Como clasificador final se seleccionó aquel obtenido de la búsqueda en cuadrícula.
Una vez construido el stack ensemble, se evaluó su desempeño utilizando la validación cruzada de 3 pliegues y midiendo la AUC.

```
# Construir stacked ensemble de árboles de decisión:
dt_stacking = DecisionTreeClassifier(random_state=42, **clf.best_params_)
estimators = [('dt', DecisionTreeClassifier(random_state=42)),
('dt_best_params', dt_stacking)]
clf_stacking = StackingClassifier(estimators=estimators, final_estimator=dt_stacking)

# Evaluar el modelo utilizando validación cruzada de 3 pliegues:
cv_results = cross_val_score(clf_stacking, X_oversampled, y_oversampled, cv=skf, scoring='roc_auc')

# Medir el desempeño del modelo
print("AUC: %.3f (%.3f)" % (np.mean(cv_results), np.std(cv_results)))
```

```
predictions = cross_val_predict(clf_stacking, X_oversampled, y_oversampled, cv=skf)

print("Accuracy: %.3f" % accuracy_score(y_oversampled, predictions))
print("Precision: %.3f" % precision_score(y_oversampled, predictions))
print("Recall: %.3f" % recall_score(y_oversampled, predictions))
print("F1 Score: %.3f" % f1_score(y_oversampled, predictions))
```
